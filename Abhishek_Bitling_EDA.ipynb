**3.Convert Date Columns to Datetime Format**

```python
# Convert date columns
customers_df["SignupDate"] = pd.to_datetime(customers_df["SignupDate"])
transactions_df["TransactionDate"] = pd.to_datetime(transactions_df["TransactionDate"])
```

**5.Analyze Sales Trends Over Time**

```python
import matplotlib.pyplot as plt

# Transactions over time
transactions_df.resample('M', on='TransactionDate').count()["TransactionID"].plot(figsize=(8,3), title="Number of Transactions Over Time")
plt.xlabel("Date")
plt.ylabel("Transaction Count")
plt.grid(True)
plt.show()
```

**6.Identify Best-Selling Products**

```python
import seaborn as sns

# Top 10 most purchased products
top_products = transactions_df["ProductID"].value_counts().head(10)

plt.figure(figsize=(8,3))
sns.barplot(x=top_products.index, y=top_products.values, palette="viridis")
plt.title("Top 10 Most Purchased Products")
plt.xlabel("Product ID")
plt.ylabel("Purchase Count")
plt.xticks(rotation=45)
plt.show()
```

**7.Revenue Contribution by Region**

```python
# Merge transactions with customer data
merged_df = transactions_df.merge(customers_df, on="CustomerID")

# Revenue by region
revenue_by_region = merged_df.groupby("Region")["TotalValue"].sum().sort_values(ascending=False)

plt.figure(figsize=(6,4))
sns.barplot(x=revenue_by_region.index, y=revenue_by_region.values, palette="coolwarm")
plt.title("Total Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Total Revenue (USD)")
plt.show()
```

**8.Customer Purchase Behavior Analysis**

```python
# Histogram of total purchase values
plt.figure(figsize=(10, 5))
sns.histplot(transactions_df["TotalValue"], bins=30, kde=True, color="blue")
plt.title("Distribution of Total Transaction Values")
plt.xlabel("Total Value (USD)")
plt.ylabel("Frequency")
plt.show()
```

**9.Identify High-Value Customers**

```python
# Top 10 customers by total spending
top_customers = merged_df.groupby("CustomerID")["TotalValue"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(6,4))
sns.barplot(x=top_customers.index, y=top_customers.values, palette="magma")
plt.title("Top 10 Customers by Total Spending")
plt.xlabel("Customer ID")
plt.ylabel("Total Spending (USD)")
plt.xticks(rotation=45)
plt.show()
```

```python
from fpdf import FPDF

class PDF(FPDF):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        # Add this line to set the font encoding to UTF-8
        self.set_font('Arial', '', 10) # Set default font and size, and encoding
        self.font_family = 'Arial' # Set default font family
        self.font_style = '' # Set default font style
        self.font_size_pt = 10 # Set default font size in points
        self.font_encoding = 'utf-8' # Set font encoding to UTF-8

    def header(self):
        self.set_font("Arial", "B", 12)
        self.cell(200, 10, "Business Insights from EDA", ln=True, align="C")

    def chapter_title(self, title):
        self.set_font("Arial", "B", 12)
        self.cell(0, 10, title, ln=True, align="L")
        self.ln(5)

    def chapter_body(self, body):
        self.set_font("Arial", "", 10)
        # Encode the body text to UTF-8 before adding it to the PDF
        self.multi_cell(0, 10, body.encode('latin-1', 'replace').decode('latin-1'))
        self.ln()

# Create PDF report
pdf = PDF()
pdf.add_page()

pdf.chapter_title("1. Seasonal Sales Trends")
pdf.chapter_body("Transaction volume shows clear fluctuations over time, indicating potential seasonality. Identifying high-sales periods can help optimize marketing campaigns, stock management, and promotional offers.")

pdf.chapter_title("2. High Revenue Generating Regions")
pdf.chapter_body("Some regions contribute significantly more revenue than others. By targeting underperforming regions with localized marketing and incentives, the company can improve overall revenue and market penetration.")

pdf.chapter_title("3. Top-Selling Products Influence Demand")
pdf.chapter_body("A small subset of products dominates sales, meaning optimizing inventory for these high-demand items can reduce stockouts and maximize profitability.")

pdf.chapter_title("4. Customer Purchase Behavior Variations")
pdf.chapter_body("The average purchase quantity per transaction is around 2.5, but high-value transactions reach nearly $2,000. This suggests different customer segmentsbudget shoppers and premium buyerswho may require different marketing strategies.")

pdf.chapter_title("5. Significant Customer Growth Potential")
pdf.chapter_body("With 199 customers making transactions but 200 total customers in the dataset, there might be inactive customers. Re-engagement campaigns could convert them into active buyers and increase retention.")

pdf.output("Abhishek_Bitling_EDA.pdf")
```
